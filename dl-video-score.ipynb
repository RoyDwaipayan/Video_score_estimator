```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
import os
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
os.listdir('../input/video-information-dataset')
```

```python
data = pd.read_csv('../input/data-preprocessed-3-target-outlier-not-removed/data_preprocessed2.csv')
data_train = data.sample(n=1000)
predictor = data_train.drop(['engagement_score'],axis=1)
n_cols_predictor = predictor.shape[1]
# print(predictor)
target = data_train['engagement_score']
```

```python
scaler = StandardScaler()
scaler.fit(predictor)
```

```python
predictor_scaled = scaler.transform(predictor)
```

```python
model = Sequential()
#model.add(Dense(25,activation='tanh',input_shape=(n_cols_predictor,)))
model.add(Dense(25,activation='tanh'))
model.add(Dense(100,activation='softmax'))
#model.add(Dense(200,activation='exponential'))
#model.add(Dense(100,activation='softplus'))
model.add(Dense(50,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
model.fit(predictor_scaled,target,epochs=2000)
```

```python
data_test = data.loc[data.index.difference(data_train.index)]
test_predictor = data_test.drop(['engagement_score'],axis=1)
```

```python
test_predictions = model.predict(test_predictor)
```

```python
test_predictions.shape
```

```python
test_target = data_test['engagement_score']
```

```python
err=np.sum(np.abs(test_predictions-test_target.values),axis=0)
real=np.sum(test_target.values,axis=0)
print("Accuracy:")
print(100-(err*100/real))
```
